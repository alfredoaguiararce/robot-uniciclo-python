### 1.  Se agregan librerias para calculos simbolicos y matematicos.

```python
# importamos las librerías necesarias
import sympy as sp  # librería para cálculo simbólico
from sympy import *
from sympy.physics.vector import init_vprinting
init_vprinting(use_latex='mathjax', pretty_print=False)
from sympy.physics.mechanics import dynamicsymbols
```

### 2.- Se definen las variables simbolicas de las ecuaciones

```python
w11, w12, w21, w22, w31, w32, w41, w42, w43, w44, w51, w52, w53, w54, w61, w62, w63, w64, w71, w72, w73, w74 = dynamicsymbols('w_11 w_12 w_21 w_22 w_31 w_32 w_41 w_42 w_43 w_44 w_51 w_52 w_53 w_54 w_61 w_62 w_63 w_64 w_71 w_72 w_73 w_74')
```

```python
#Imprimimos las variables
w11, w12, w21, w22, w31, w32, w41, w42, w43, w44, w51, w52, w53, w54, w61, w62, w63, w64, w71, w72, w73, w74
```

```python
Sx11, Sx12, Sx21, Sx22, Sx31, Sx32, G1, G2, u11, u12 = dynamicsymbols('S(x_11) S(x_12) S(x_21) S(x_22) S(x_31) S(x_32) G_1 G_2 u_11 u_12')
```

```python
#Imprimimos las variables
Sx11, Sx12, Sx21, Sx22, Sx31, Sx32, G1, G2, u11, u12
```

### 3. Definimos las ecuaciones simbolicas expuestas en el articulo cientifico, ecuaciones (22).

image.png

```python
x1 = (w11*Sx11)+(w12*Sx12)
x2 = (w21*Sx11)+(w22*Sx12)
x3 = (w31*Sx11)+(w32*Sx12)
x4 = (w41*Sx11)+(w42*Sx12)+(w43*Sx21)+(w44*Sx31)
x5 = (w51*Sx11)+(w52*Sx12)+(w53*Sx21)+(w54*Sx31)+(G1*u11)
x6 = (w61*Sx11)+(w62*Sx12)+(w63*Sx22)+(w64*Sx32)
x7 = (w71*Sx11)+(w72*Sx12)+(w73*Sx22)+(w74*Sx32)+(G2*u12)
```

### 4.- Matrices de medicion 

H ϵ R ^ (m x L)

* L =  total de pesos.
* m = numero de salidas.

```python
"""
  Para X1 : 
    L = 2
    m = 1
"""
H1 = [diff(x1,w11), diff(x1,w12)]

"""
  Para X2 : 
    L = 2
    m = 1
"""
H2 = [diff(x2,w21), diff(x2,w22)]

"""
  Para X3 : 
    L = 2
    m = 1
"""
H3 = [diff(x3,w31), diff(x3,w32)]

"""
  Para X4 : 
    L = 4
    m = 1
"""
H4 = [diff(x4,w41), diff(x4,w42), diff(x4,w43), diff(x4,w44) ]

"""
  Para X5 : 
    L = 4
    m = 1
"""
H5 = [diff(x5,w51), diff(x5,w52), diff(x5,w53), diff(x5,w54) ]

"""
  Para X6 : 
    L = 4
    m = 1
"""
H6 = [diff(x6,w61), diff(x6,w62), diff(x6,w63), diff(x6,w64) ]

"""
  Para X7 : 
    L = 4
    m = 1
"""
H7 = [diff(x7,w71), diff(x7,w72), diff(x7,w73), diff(x7,w74) ]
```
